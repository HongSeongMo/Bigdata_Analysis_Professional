```python
import pandas as pd
data = pd.DataFrame({'학번':['2301','2302','2303'],
                     '성명':['홍길동','장유진','임영미'],
                     '나이':[27,19,35],
                     '영어':[70,85,75],
                     '수학':[80,80,90]})
data
```

```python
df = data.set_index('학번')
df
```

```python
print(df.iterrows())
```

```python
df.iterrows()
```

```python
sumeng = 0
for index, row in df.iterrows():
    sumeng += row['영어']
    print(index, row['성명'], row['영어'])
print(sumeng)
```

```python
summath = 0
for i in range(0, df.shape[0]):
    index = df.index[i]
    summath += df.iloc[i][3]
    print(index, df.loc[index]['성명'], df.loc[index]['수학'])
print(summath)
```

```python
sumhap = 0
for col in df.columns:
    if col == '영어' or col == '수학':
        sumhap += df[col]
print(sumhap)
```

```python
sumhap = 0
for i in range(0, df.shape[1]):
    col = df.columns[i]
    if col == '영어' or col == '수학':
        sumhap += df.iloc[:,i]
print(sumhap)
```

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/work/usedcars.csv', encoding='euc-kr', index_col=0)
df.head()
```

```python
dfnew = df[['maker','fuel']]
dfnew.head()
```

```python
crosstable = pd.crosstab(dfnew.maker, dfnew.fuel, margins=True)
crosstable
```

```python
print(crosstable.loc['현대',['디젤']])
```

```python
maker_freq = dfnew['maker'].value_counts()
maker_freq
```

```python
make_prop = dfnew['maker'].value_counts(normalize=True)*100
make_prop
```

```python
result_df = pd.concat([maker_freq, make_prop], axis=1)
result_df.columns = ['Frequency','Ratio(%)']
result_df
```
