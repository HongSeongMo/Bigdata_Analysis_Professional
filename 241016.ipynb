```python
import pandas as pd
data = pd.DataFrame({'학번':['2301','2302','2303'],
                     '성명':['홍길동','장유진','임영미'],
                     '나이':[27,19,35],
                     '영어':[70,85,75],
                     '수학':[80,80,90]})
data
```

```python
df = data.set_index('학번')
df
```

```python
print(df.iterrows())
```

```python
df.iterrows()
```

```python
sumeng = 0
for index, row in df.iterrows():
    sumeng += row['영어']
    print(index, row['성명'], row['영어'])
print(sumeng)
```

```python
summath = 0
for i in range(0, df.shape[0]):
    index = df.index[i]
    summath += df.iloc[i][3]
    print(index, df.loc[index]['성명'], df.loc[index]['수학'])
print(summath)
```
