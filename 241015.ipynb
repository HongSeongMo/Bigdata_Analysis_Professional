```python
air = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/work/airquality.csv', index_col=0)
air.head()
```

```python
data = air.dropna(subset = ['Ozone'])
data.describe()
```

```python
iqr = data['Ozone'].describe()['75%'] - data['Ozone'].describe()['25%']
iqr
```

```python
r1 = data['Ozone'].mean() - 0.5 * iqr
r1
```

```python
r2 = data['Ozone'].mean() + 0.5 * iqr
r2
```

```python
cond = (data['Ozone'] <= r1) | (data['Ozone'] >= r2)
cond.head()
```

```python
print(sum(cond))
```

```python
print(data[cond].sum())
```

```python
print(data[cond].mean())
```

```python
print(data[cond]['Solar.R'].mean())
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

```python
air = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/work/airquality.csv', index_col=0)
air.head()
```

```python
data = air.dropna()
data.head()
```

```python
zscale = lambda x: (x - x.mean()) / x.std()
data['Ozone'] = zscale(data['Ozone'])
data.head()
```
