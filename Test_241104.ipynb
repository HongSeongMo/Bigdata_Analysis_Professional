```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/work/mtcars.csv', header=0, index_col=0)
df.head()
```

```python
 dfnew = df[df['am'] == 0]
 dfnew.head()
```

```python
q1 = dfnew['mpg'].quantile(0.25)
q3 = dfnew['mpg'].quantile(0.75)
iqr = q3 - q1
```

```python
avg_mpg = dfnew.mpg.mean()
avg_mpg
```

```python
cond = (dfnew['mpg'] >= avg_mpg + iqr) | (dfnew['mpg'] <= avg_mpg - iqr)
cnt = dfnew[cond].am.count()
print(cnt)
```

```python
avg = dfnew[cond].mpg.mean()
print(avg)
```
