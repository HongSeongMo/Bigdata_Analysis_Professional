```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/work/data.csv', encoding='euc-kr', index_col=0)
df.head()
```

```python
print('"성별"의 데이터 유형: ', end='')
print(df.dtypes['성별'])
```

```python
df['gender'] = df['성별'].astype('category')
print(df.dtypes['gender'])
```

```python
df['gendernum'] = df['gender'].map({'남자':1,'여자':0})
df['gendernum'] = df['gendernum'].astype(int)
print(df['gendernum'].dtypes)
```

```python
df.gendernum.count()
```

```python
df['gendernum'].sum()
```

```python
len(df.loc[df['gendernum'] == 0])
```

```python
df.loc[df['gendernum'] == 0].shape[0]
```

```python
ratio_male = df['gendernum'].sum()/df.gendernum.count() * 100
print(ratio_male,'%')
```

```python
ratio_female = len(df.loc[df['gendernum'] == 0])/df.shape[0] * 100
print(ratio_female,'%')
```
