```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/work/data.csv', encoding='euc-kr', index_col=0)
data.head(2)
```

```python
df = data[['쇼핑1월', '쇼핑2월', '쇼핑3월']]
df.head()
```

```python
def customer(x):
    if x['쇼핑2월'] >= 50 and x['쇼핑3월'] >= 52:
        return '우수고객'
    else:
        return '일반고객'

df['isvip'] = df.apply(customer, axis=1)
df.head()
```

```python
df - data[['쇼핑1월', '쇼핑2월', '쇼핑3월']]
df.head()
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/work/data.csv', encoding='euc-kr', index_col=0)
data.head(2)
```

```python
df = data[['쇼핑1월', '쇼핑2월', '쇼핑3월']]
df.head()
```

```python
df['sum'] = df.apply(np.sum, axis=1)
df['avg'] = df.apply(np.mean, axis=1)
df.head()
```

```python
def gap(x):
    return (x['쇼핑2월'] + x['쇼핑3월'])/2

df['(2월, 3월) 평균'] = df.apply(gap, axis=1)
df.head()
```
