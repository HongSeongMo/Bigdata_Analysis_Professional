```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/work/mtcars.csv', index_col=0)
data.head()
```

```python
df = data[['mpg', 'wt' ,'qsec', 'disp', 'drat']]
df.head()
```

```python
df.cov()
```

```python
df.corr()
```

```python
for i in range(0, df.shape[1]):
    col1 = df.columns[i]
    j = i + 1
    for k in range(j, df.shape[1]):
        col2 = df.columns[k]
        result = stats.pearsonr(df[col1], df[col2])
        print('변수1: ' + df.columns[i] +' '+ '변수2: ' + df.columns[k] + ' (검정통계량, 유의확률)= ', end='');print(result)
```
