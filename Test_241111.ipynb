```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
air = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/work/airquality.csv', index_col=0)
air.head()
```

```python
cond = (air['Month'] == 8) & (air['Day'] == 26)
data = air[air['Month'] == 8]
data
```

```python
d1 = air[cond].Ozone
d2 = air[cond]['Solar.R']
print('8월 26일 오존량: ', d1)
print('8월 26일 태양 복사량: ', d2)
```

```python
print(type(d1))
```

```python
check1 = data['Ozone'] >= float(d1)
check2 = data['Solar.R'] >= float(d2)
print(sum(check1))
print(sum(check2))
```

```python
print(data[check1].Ozone.mean())
print(data[check2]['Solar.R'].mean())
```
