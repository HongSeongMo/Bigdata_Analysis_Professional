```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns = iris['feature_names'] + ['target'])
df.head()
```

```python
dfnew = df[df['target'] != 0]
dfnew
```

```python
m = dfnew['sepal length (cm)'].median()
n = dfnew['petal length (cm)'].median()
df1 = dfnew[dfnew['target'] == 1]
df2 = dfnew[dfnew['target'] == 2]
p = len(df1[df1['sepal length (cm)'] >= m])/df1.shape[0]
q = len(df2[df2['petal length (cm)'] >= n])/df2.shape[0]
print(m,n)
print(df1.head(), df2.head())
print(p,q)
print(q - p)
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns = iris['feature_names'] +['target'])
df.head()
```

```python
dfnew = df[df['target'] == 0]
dfnew.tail()
```

```python
m = dfnew['sepal length (cm)'].median()
print(m)
```

```python
dfnew['value'] = np.where(dfnew['sepal length (cm)'] > m, 1, 0)
df1 = dfnew[dfnew['value'] == 1]
df1['sepal length (cm)'].mean()
```
