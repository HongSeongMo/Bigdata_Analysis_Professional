# iris 데이터세트 활용 작업형 문제
Q-01

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
iris = load_iris()
irisdata = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns = iris['feature_names'] + ['target'])
irisdata.head()
```

```python
print('붓꽃 꽃잎 길이의 (평균+표준편차)')
print(irisdata['petal length (cm)'].mean() + irisdata['petal length (cm)'].std())
print(irisdata.describe())
print(irisdata.describe()['petal length (cm)']['mean'] + irisdata.describe()['petal length (cm)']['std'])
print('붓꽃 꽃잎 너비의 (평균+표준편차)')
print(irisdata['petal width (cm)'].mean() + irisdata['petal width (cm)'].std())
```

```python
dfnew = irisdata[irisdata['target'] == 0]
print(dfnew.head())
x = dfnew['petal length (cm)']
dfnew['minmax-scale'] = (x - x.min(axis=0)) / (x.max(axis=0) - x.min(axis=-0))
print(dfnew.head())
print(dfnew['minmax-scale'].mean())
dfnew['zscore-scale'] = (x - x.mean(axis=0)) / x.std(axis=0)
print(dfnew.head())
print(dfnew['zscore-scale'].mean())
print(dfnew.describe())
```

```python
minmax1 = MinMaxScaler()
d1 = minmax1.fit_transform(dfnew)
dfnew['minmax_scale1'] = d1[:,2]
# d2 = minmax_scale(dfnew)
# dfnew['minmax_scale2'] = d2[:,2]
print(dfnew.head())
print(dfnew['minmax_scale1'].mean())
print(dfnew['minmax_scale2'].mean())
```
