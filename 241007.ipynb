```python
import pandas as pd
import datetime as dt
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/air_pollution_data.csv', encoding='euc-kr', index_col=0, header=0)

pd.options.display.max_columns = None
pd.options.display.max_colwidth = 30
df
```

```python
df.dtypes
```

```python
df['측정일자'] = df['dataTime'].apply(lambda x: pd.to_datetime(str(x), format ='%Y-%m-%d'))
df.dtypes
```

```python
df['연도'] = df['측정일자'].dt.year
df
```

```python
df['월'] = df['측정일자'].dt.month
df['일'] = df['측정일자'].dt.day
df
```

```python
df.head()
```

```python
cond = (df['연도'] == 2022) & (df['월'] == 6)
df[cond]
```

```python
np.mean(df[cond]['coValue'])
```

```python
np.mean(df[cond].coValue)
```
